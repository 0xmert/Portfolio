```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import dateutil.parser as parser
import altair as alt
import csv
```

```python
## Import the dataset
url = 'TimeSeries.csv'
dusaf_df = pd.read_csv(url)
```

```python
### Transpoze the dataset
dusaf_T= dusaf_df.T
## Check the types
print(dusaf_T.dtypes)
```

```python
## After the Transpose operation, I have to remove first, and last 2 rows.
Last_2rows = dusaf_T.tail(2)
First_row = dusaf_T.head(1)
print(Last_2rows)
print(First_row)
dusaf_T.drop(dusaf_T.tail(2).index,
        inplace = True)
dusaf_T.drop(dusaf_T.head(1).index,
        inplace = True)
print(dusaf_T)
```

```python
First_row
```

```python
## Copy the dataframe, and add dates as new column
Date_as_column = dusaf_T.reset_index(level=0)
index = 0
for row in dusaf_T.itertuples():
    Date_as_column.at[index,'index']= parser.parse(row[0])
    index += 1
```

```python
## Let's check how it looks
Date_as_column
```

```python
## Extract each field's own time series, save it in a list
list_fields = []
for i in range(1028):
    field_id = 'field_' + str(i)  
    field_id = Date_as_column[(Date_as_column[i] > -9999)]
    field_id.drop(field_id.columns.difference(['index',i]), 1, inplace=True)
    list_fields.append(field_id)
```

```python
list_fields[1027] = list_fields[1027].reset_index(drop=True)
```

```python
## Change the column names into string
source = list_fields[0]
source.columns = source.columns.astype(str)
source['index'] = source['index'].astype(str)
source
```

```python
## Go over every row in the column
## Calculate the distance between consecutive elements
## If there's a big step, save the date.
```

```python
for i in range(1027):
    if len(list_fields[i]) == 0:
        list_fields.pop(i)
```

```python
list_fields[72]
```

```python
Dictionary_all = {}
for field_number in range(1028):
    ## Reset the index of each table first
    list_fields[field_number] = list_fields[field_number].reset_index(drop=True)
    iteration_number_innerLoop = len(list_fields[field_number]) - 2
    ## Create a temporary dictionary for the given field.
    Dictionary_Temporary = {}
    ## if the field is 73, skip it cause there's no data about it. Need to find a better solution for this later.
    if i != 73:
        for row in range(iteration_number_innerLoop):
            ## Check consecutive days
            consecutive_1 = list_fields[field_number][field_number][row]
            consecutive_2 = list_fields[field_number][field_number][row+1]
            abs_ = abs(consecutive_1 - consecutive_2)
            if abs_ > 0.5:
                ## Assign all the days that increased above the threshold into a temporary dictionary.
                Dictionary_Temporary[str(list_fields[field_number]['index'][row]) + "__fid: "+ str(field_number)] = consecutive_1
        ## At the end of for loop, assign the recorded dictionary into a global dictionary.
        Dictionary_all[field_number] = Dictionary_Temporary
print(Dictionary_all)
```

```python
with open('fielddict.csv', 'w') as f:  # You will need 'wb' mode in Python 2.x
    w = csv.DictWriter(f, Dictionary_all.keys())
    w.writeheader()
    w.writerow(Dictionary_all)
```
